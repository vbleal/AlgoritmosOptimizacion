## **3. Algoritmos con Vuelta Atr치s (*Backtracking*)**

## **$N$ Reinas**

```python
#N Reinas - Vuelta Atr치s()
################################################################################

#Verifica que en la soluci칩n parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la soluci칩n tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " est치 " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la soluci칩n al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]

  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

reinas(8,solucion=[],etapa=0)
```

## **Viaje por el R칤o**

```python
#Viaje por el rio - Programaci칩n din치mica
################################################################################

TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

#999 se puede sustituir por float("inf")


#Calculo de la matriz de PRECIOS y RUTAS
################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicializaci칩n de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################

PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

## **Puntos m치s cercanos**
---

### 游닆**Planteamiento del Problema**

*  **Supuesto**:
   *  Dado un conjunto de puntos, el objetivo es encontrar los dos puntos m치s cercanos.

*  **Gu칤a para aprendizaje**:

   1. Suponer en 1D, es decir, una lista de n칰meros, como **`[3403, 4537, 9089, 9746, 7259, ...]`**.

   2. Primer intento: Utilizar el m칠todo de fuerza bruta para encontrar la pareja de puntos m치s cercanos.

   3. Calcular la complejidad del m칠todo de fuerza bruta y considerar si se puede mejorar.

   4. Segundo intento: Aplicar el m칠todo Divide y Vencer치s para intentar mejorar la complejidad del algoritmo.

   5. Calcular la complejidad del m칠todo Divide y Vencer치s y determinar si hay mejoras.

   6. Extender el algoritmo a 2D, trabajando con puntos definidos por pares de n칰meros, como **`[(1122, 6175), (135, 4076), (7296, 2741)...]`**.

   6. Extender el algoritmo a 3D.

### 游빑**An치lisis**

#### **1D**

*  **1D con Fuerza Bruta**

Para la lista en 1D, el m칠todo de fuerza bruta implicar칤a comparar cada punto con todos los dem치s puntos para encontrar la distancia m칤nima.

*  **1D con Divide y Vencer치s**

En 1D, este algoritmo implicar칤a dividir la lista de n칰meros ordenada en dos mitades, encontrar el par de puntos m치s cercanos dentro de cada mitad y luego encontrar el par m치s cercano que cruza el punto medio.

#### **2D**

En 2D, adem치s de dividir los puntos, tendr칤a que considerarse un 치rea alrededor de la l칤nea divisoria donde los puntos m치s cercanos podr칤an residir en diferentes mitades.

#### **3D**

En 3D, lo anterior se extiende a un volumen alrededor del plano divisorio.

### 九勇**Soluci칩n**

```python
import math
```

#### **1D**

*  **1D con Fuerza Bruta**

```python
# 1D: Fuerza Bruta
def closest_pair_brute_force(points):
    min_distance = float('inf')
    closest_points = None
    for i in range(len(points) - 1):
        for j in range(i + 1, len(points)):
            distance = abs(points[i] - points[j])
            if distance < min_distance:
                min_distance = distance
                closest_points = (points[i], points[j])
    return closest_points, min_distance
```

La complejidad computacional de este enfoque es $O(n)$, donde $n$ es el n칰mero de puntos en la lista. Ahora, se proceder치 a intentar mejorar esta complejidad utilizando el enfoque de Divide y Vencer치s.

Es decir:
1. Se itera sobre cada punto de la lista.

2. Para cada punto, se itera nuevamente sobre todos los otros puntos para comparar las distancias.

3. Esto significa que para cada uno de los $n$ puntos, se hacen $n - 1$ comparaciones (en la pr치ctica, es aproximadamente $n$ comparaciones porque $n - 1$ es casi igual a $n$ para grandes valores de $n$).


Esto da lugar a un bucle anidado, donde la cantidad de operaciones es proporcional a $n\cdot n$, que se simplifica como $O(n)$. Este es un crecimiento cuadr치tico, lo que significa que si se duplica el n칰mero de puntos, el tiempo que toma resolver el problema se multiplica por $4$.

*  **1D con Divide y Vencer치s**

```python
# 1D: Divide y Vencer치s
def closest_pair_divide_and_conquer(points):
    if len(points) <= 3:
        return closest_pair_brute_force(points)
    mid = len(points) // 2
    left_points = points[:mid]
    right_points = points[mid:]
    left_pair, left_distance = closest_pair_divide_and_conquer(left_points)
    right_pair, right_distance = closest_pair_divide_and_conquer(right_points)
    closest_pair, closest_distance = (left_pair, left_distance) if left_distance < right_distance else (right_pair, right_distance)
    return closest_pair, closest_distance
```

La complejidad computacional de este m칠todo es $O(n\cdot \mathrm{log}(n))$, debido principalmente al paso de ordenamiento. La parte de dividir y vencer realmente opera en $O(n)$, pero como se necesita ordenar la lista al principio, esa ser칤a la complejidad dominante.

N칩tese que:

1. Antes de aplicar el m칠todo, la lista de puntos se ordena, lo cual tiene una complejidad temporal de $O(n\cdot \mathrm{log}(n))$ para algoritmos de ordenamiento eficientes como ***Merge Sort*** o ***Quick Sort***.

2. El m칠todo divide y vencer치s divide la lista de puntos en mitades hasta llegar a un caso base que se resuelve con fuerza bruta. Esto sigue una progresi칩n logar칤tmica, dividiendo el conjunto de datos en cada paso. El n칰mero de divisiones que se pueden hacer es proporcional a $\mathrm{log}(n)$.

3. En cada nivel de divisi칩n, se hacen comparaciones que en total suman n operaciones (por ejemplo, comparar los puntos m치s cercanos en la franja central). Debido a que esto se hace en cada nivel del 치rbol de recursi칩n, se multiplica por $\mathrm{log}(n)$, lo que da una complejidad de las operaciones de divisi칩n y conquista de $O(n\cdot \mathrm{log}(n))$.

La raz칩n por la cual la complejidad dominante es $O(n\cdot \mathrm{log}(n))$ y no solo $O(n)$, a pesar de que el divide y vencer치s por s칤 solo ser칤a $O(n)$, es porque el paso de ordenamiento es necesario y su complejidad no puede ser ignorada. Es el paso m치s costoso y, por lo tanto, define la complejidad general del algoritmo.

#### **2D**

*  **Distancia Euclideana en 2D**

```python
# Distancia Euclideana en 2D
def euclidean_distance_2d(point1, point2):
    return math.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)
```

*  **2D con Fuerza Bruta**

```python
# 2D con Fuerza Bruta
def closest_pair_brute_force_2d(points):
    min_distance = float('inf')
    closest_points = None
    for i in range(len(points) - 1):
        for j in range(i + 1, len(points)):
            distance = euclidean_distance_2d(points[i], points[j])
            if distance < min_distance:
                min_distance = distance
                closest_points = (points[i], points[j])
    return closest_points, min_distance
```

*  **2D con Divide y Vencer치s**

```python
# 2D con Divide y Vencer치s
def closest_pair_dc_2d(points):
    if len(points) <= 3:
        return closest_pair_brute_force_2d(points)

    mid = len(points) // 2
    left_points = points[:mid]
    right_points = points[mid:]
    left_pair, left_distance = closest_pair_dc_2d(left_points)
    right_pair, right_distance = closest_pair_dc_2d(right_points)
    closest_pair, closest_distance = (left_pair, left_distance) if left_distance < right_distance else (right_pair, right_distance)

    strip_points = [point for point in points if abs(point[0] - points[mid][0]) < closest_distance]
    strip_points.sort(key=lambda point: point[1])

    for i in range(len(strip_points)):
        for j in range(i+1, len(strip_points)):
            if strip_points[j][1] - strip_points[i][1] < closest_distance:
                distance = euclidean_distance_2d(strip_points[i], strip_points[j])
                if distance < closest_distance:
                    closest_distance = distance
                    closest_pair = (strip_points[i], strip_points[j])
            else:
                break

    return closest_pair, closest_distance
```

#### **3D**

*  **Distancia Euclideana en 3D**

```python
# Distancia Euclideana
def euclidean_distance_3d(point1, point2):
    return math.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2 + (point1[2] - point2[2])**2)
```

*  **3D con Fuerza Bruta**

```python
# 3D con Fuerza Bruta
def closest_pair_brute_force_3d(points):
    min_distance = float('inf')
    closest_points = None
    for i in range(len(points) - 1):
        for j in range(i + 1, len(points)):
            distance = euclidean_distance_3d(points[i], points[j])
            if distance < min_distance:
                min_distance = distance
                closest_points = (points[i], points[j])
    return closest_points, min_distance
```

*  **3D con Divide y Vencer치s**

```python
# 3D con Divide y Vencer치s
def closest_pair_dc_3d(points):
    if len(points) <= 3:
        return closest_pair_brute_force_3d(points)

    mid = len(points) // 2
    left_points = points[:mid]
    right_points = points[mid:]
    left_pair, left_distance = closest_pair_dc_3d(left_points)
    right_pair, right_distance = closest_pair_dc_3d(right_points)
    closest_pair, closest_distance = (left_pair, left_distance) if left_distance < right_distance else (right_pair, right_distance)

    strip_points = [point for point in points if abs(point[0] - points[mid][0]) < closest_distance]
    strip_points.sort(key=lambda point: (point[1], point[2]))

    for i in range(len(strip_points)):
        for j in range(i+1, min(i + 7, len(strip_points))):
            distance = euclidean_distance_3d(strip_points[i], strip_points[j])
            if distance < closest_distance:
                closest_distance = distance
                closest_pair = (strip_points[i], strip_points[j])

    return closest_pair, closest_distance
```
